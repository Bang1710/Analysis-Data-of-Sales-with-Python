## Analysis Data of Sales with Python

### I. About the Dataset

#### 1.The dataset consists of 11 columns, each column representing an attribute of purchase on a product
##### - Order ID: A unique ID for each order placed on a product
##### - Product: Item that is purchased
##### - Quantity Ordered: Describes how many of that products are ordered
##### - Price Each: Price of a unit of that product
##### - Order Date: Date on which the order is placed
##### - Purchase Address: Address to where the order is shipped
##### - Month, Sales, City, Hour: Extra attributes formed from the above.

### Once we have cleaned up our data a bit, we move the data exploration section. In this section we explore 11 high level business questions related to our data.

##### 1. What is the most ordered product in the store ?

```python
product_quantity = data.groupby('Product')['Quantity Ordered'].sum()
ax = product_quantity.plot.bar(title = 'Quantity ordered per product',color='#2CA02C',figsize =(16,9))
# t nhn trc x v trc y
ax.set_xlabel('Quantity Ordered')
ax.set_ylabel('Product')
plt.show()
```

##### 2. What is the relationship between the average selling price of products and the quantity of orders placed ?

```python
# Tnh tng 'Quantity Ordered' theo sn phm
product_quantity = data.groupby('Product')['Quantity Ordered'].sum()

# Tnh gi trung bnh ('Price Each') theo sn phm
prices = data.groupby('Product')['Price Each'].mean()

# Danh sch tn sn phm
products = product_quantity.index

# V biu 
fig, ax1 = plt.subplots(figsize=(16, 9))

# V ct u tin: 'Quantity Ordered'
ax1.bar(products, product_quantity, color='g', label='Quantity Ordered')
ax1.set_ylabel('Quantity Ordered', color='g')

# To mt twin axes cho trc y th hai
ax2 = ax1.twinx()

# V ct th hai: 'Price (mean)'
ax2.plot(products, prices, color='b', label='Price (mean)')
ax2.set_ylabel('Price ($)', color='b')

# t nhn trc x v tiu 
ax1.set_xlabel('Product')
# ax1.set_xticklabels(products)
plt.title('Quantity Ordered and Price (mean) per product')

# Hin th legend
fig.legend(loc='upper right')
plt.show()
```

##### 3. What are the top 5 cities with the highest number of orders placed?

```python
city_sales = data.groupby('City')['Sales'].sum().nlargest(5)
city_sales
```

```python
plt.figure(figsize = (8,8))
plt.title('Top 5 citys with maximum sales')
plt.pie(city_sales,autopct = '%.2f',labels = city_sales.index)
plt.show()
```

```python
# Thit lp kiu style
sns.set_style("whitegrid")

# V biu 
sns.barplot(x=city_sales.index, y=city_sales)

# t tn biu 
plt.title('top 5 city with most sales')

# Hin th biu 
plt.show()
```

##### 4. What is the number of products ordered per hour?

```python
data.groupby('Hour')['Sales'].sum().plot.bar(title = 'Per hour sales',figsize = (16, 9))
```

```python
# Add hour column
data['Hour'] = pd.to_datetime(data['Order Date']).dt.hour
data['Minute'] = pd.to_datetime(data['Order Date']).dt.minute
data['Count'] = data['Quantity Ordered']
data.head()
```

##### 5. What time should we display advertisemens to maximize the likelihood of customers buying product ?

```python
# To danh sch cc gi
keys = sorted(data['Hour'].unique())

# V biu 
plt.plot(keys, data.groupby(['Hour']).count()['Count'])
plt.xticks(keys)
plt.grid(True)
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.show()
# My recommendation is slightly before 11am or 7pm
```

##### 6. What is the total sales revenue of the store for each month ?

```python
# Tnh tng doanh s bn hng trong mi thng
monthly_sales = data.groupby('Month')['Sales'].sum()

# In ra kt qu
monthly_sales
```

##### 7. Which product is the best-selling?

```python
product_quantity = data.groupby('Product')['Quantity Ordered'].sum()

# Sn phm c s lng t hng cao nht
most_ordered_product = product_quantity.idxmax()
most_ordered_product
```

##### 8. What is the highest selling price for a product ?

```python
product_price = data.groupby('Price Each')['Quantity Ordered'].sum()

# Mc gi ca sn phm c bn nhiu nht
most_ordered_price = product_price.idxmax()
most_ordered_price
```

##### 9. Is there a correlation between the price of a product and the quantity of orders placed ? 

```python
# To biu  scatter plot vi mu sc phn bit cc sn phm
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price Each', y='Quantity Ordered', hue='Product', data=data, palette='Set1')

# Thm trendline
sns.regplot(x='Price Each', y='Quantity Ordered', data=data, scatter=False)

plt.xlabel('Price Each ($)')
plt.ylabel('Quantity Ordered')
plt.title('Relationship between Price and Quantity Ordered by Product')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

##### 10. How is the distribution of prices among products in the store ? Is there a significant difference in prices among different product types ?

```python
# Phn phi gi ca cc sn phm trong ca hng
# Distribution of prices among products
plt.figure(figsize=(10, 6))
plt.hist(data['Price Each'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Product Prices')
plt.show()

# Difference in prices among different product types
plt.figure(figsize=(12, 6))
sns.boxplot(x='Product', y='Price Each', data=data)
plt.xticks(rotation=90)
plt.xlabel('Product')
plt.ylabel('Price ($)')
plt.title('Price Distribution across Product Types')
plt.show()
```

##### 11. What are the top 5 cities with the highest online order rate ?

```python
# Tnh t l t hng trc tuyn cho mi thnh ph
online_order_rate = data.groupby('City')['Count'].sum() / data.groupby('City')['Count'].count()

# Sp xp thnh ph theo t l t hng trc tuyn gim dn
highest_online_order_cities = online_order_rate.sort_values(ascending=False)

# Hin th top 5 thnh ph c t l t hng trc tuyn cao nht
print("City with top 5 online order rate:", highest_online_order_cities.head(5))
```
